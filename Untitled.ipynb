<b>Модуль А. Предобработка количественных данных

Второй способ:
Воспользоваться данными открытого источника - веб-сервиса РосНавык (https://rosnavyk.ru/). Для получения данных с этого источника можно воспользоваться вкладкой “Данные”, сделать запрос нужных вакансий и получить готовую таблицу в формате csv или xlsx.


```python
import pandas as pd

df = pd.read_csv("query.csv")
```

```python
df
```

В таблице получились пустые столбцы, удалим их

```python
all_records = df.shape[0]
df.isnull().sum()/all_records * 100
```

```python
cols = list()
for column in df.columns:
    if df[column].isnull().sum() / all_records * 100 > 42:
        cols.append(column)
print(f"Невосстановимые признаки, у которых пропусков более 60%: {cols}")
df.drop(columns=cols, inplace=True)
```

```python
df.info()
```

```python
# Приведение всех строковых значений к нижнему регистру
df[df.select_dtypes(include='object').columns] = df.select_dtypes(include='object').apply(lambda x: x.str.lower())
```

```python
df
```

В столбце Ключевые навыки много пропущенных значений, исправим это

```python
most_common_value = df['Ключевые навыки'].mode()[0]

# Заполнить пропуски в столбце 'Name' наиболее часто встречающимся значением
df['Ключевые навыки'] = df['Ключевые навыки'].fillna(most_common_value)
```

```python
df.info()
```

Изучим столбец заработная плата на наличие не подходящей информации, такой как лишний текст, вместо указания конкретной зарплаты

```python
df['Заработная плата'].unique()
```

Можно заметить не понятный текст, вместо указания зарплаты, ударим его

```python
df = df[df['Заработная плата'].apply(lambda x: len(str(x)) <= 7)]
```

```python
df['Заработная плата'].unique()
```

Изучим столбец дата публикации на наличие не подходящей информации, такой как лишний текст, вместо указания конкретной зарплаты

```python
df['Дата публикации'].unique()
```

Оставим только год, месяц и число

```python
df['Дата публикации'] = df['Дата публикации'].str[:10]
```

```python
df.to_csv('out.csv', index=False) 
```

```python
df['Заработная плата'] = pd.to_numeric(df['Заработная плата'], errors='coerce')
df['ID региона'] = pd.to_numeric(df['ID региона'], errors='coerce')
```

```python
import matplotlib.pyplot as plt

# Группировка данных по столбцу "id региона" и вычисление 
mean_salary_by_region = df.groupby('ID региона')['Заработная плата'].mean()

# Построение графика
plt.figure(figsize=(12, 6))
plt.bar(mean_salary_by_region.index, mean_salary_by_region.values, color='skyblue')
plt.xlabel('ID региона')
plt.ylabel('Средняя заработная плата')
plt.title('Зависимость средней заработной платы от id региона')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

можно заметить закономерность, как зарплата завист от определенного региона. 
В краснодарском крае зарплата больше всего

```python
# Преобразование столбца 'Дата публикации' к типу datetime
df['Дата публикации'] = pd.to_datetime(df['Дата публикации'])

# Подсчет количества публикаций по каждой дате
publications_count = df.groupby('Дата публикации').size()

# Построение диаграммы
plt.figure(figsize=(10, 6))
plt.bar(publications_count.index, publications_count.values)

plt.xlabel('Дата публикации')
plt.ylabel('Количество публикаций')
plt.title('Количество публикаций по дате публикации')

plt.xticks(rotation=45)
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

# Преобразование столбца "Работодатель" в числовой формат с помощью Label Encoding
label_encoder = LabelEncoder()
df['Работодатель_encoded'] = label_encoder.fit_transform(df['Работодатель'])

# Рассчет корреляции Пирсона между преобразованным столбцом "Работодатель" и столбцом "Заработная плата"
correlation = df['Работодатель_encoded'].corr(df['Заработная плата'])

print(f"Корреляция между столбцами 'Работодатель' и 'Заработная плата': {correlation}")
```
